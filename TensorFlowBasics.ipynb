## Many Operations

```python
#One Dimensional Gaussian curve

mean = 0.0
sigma = 1.0

z = (tf.exp(tf.negative(tf.pow(x-mean,2.0)/
                        (2.0 * tf.pow(sigma,2.0)))) *
     (1.0/(sigma*tf.sqrt(2.0*3.14))))
```

```python
import matplotlib.pyplot as plt
result = z.eval()
%matplotlib inline
plt.plot(result)
```

## Convolution

## Creating a 2-D Gaussian Kernel

```python
kernel_size = z.get_shape().as_list()[0]

print (kernel_size)

z_2d = tf.matmul(tf.reshape(z,[kernel_size,1]),tf.reshape(z,[1,kernel_size]))

plt.imshow(z_2d.eval())
```

## Convolving an Image with a Gaussian

```python
from skimage import data
img = data.camera().astype(np.float32)
plt.imshow(img, cmap='gray')
print(img.shape)
```

For image convolution in Tensorflow, images should be in 4 dimensional. 

N x H x W x C

where N-Number of Images, H-Height, W-Width, C-Number of channels.

To perform convolution with tensorflow, we have to reshape our gaussian kernel in 4 dimension as well

Kernel Height x Kernel Width x Number of Input Channels x Number of Output Channels



```python
img_4d = tf.reshape(img, [1, img.shape[0], img.shape[1],1])
print(img_4d)

print (img_4d.get_shape().as_list())

##Reshape 2D kernel into 4d format

z_4d = tf.reshape(z_2d,[kernel_size,kernel_size,1,1])
print(z_4d.get_shape().as_list())
```

## Convolve/Filter an image using a Gaussian Kernel

```python
convolved = tf.nn.conv2d(img_4d, z_4d, strides = [1,1,1,1],padding = 'SAME')
result = convolved.eval()
print (result.shape)
```
